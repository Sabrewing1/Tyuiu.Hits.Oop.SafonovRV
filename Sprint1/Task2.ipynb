````
using System;

public class Animal
{
    public int Age { get; set; }
    public string Species { get; set; }
    public static int AnimalCount { get; private set; } = 0;

    public Animal(int age, string species)
    {
        Species = species;
        Age = age;
        AnimalCount++;
    }
    
    public virtual void Eat()
    {
        Console.WriteLine($"üêæ {Species} –∫—É—à–∞–µ—Ç...");
    }

    public override string ToString()
    {
        return $"–í–∏–¥: {Species}, –í–æ–∑—Ä–∞—Å—Ç: {Age} –ª–µ—Ç";
    }
}

public class Mammal : Animal
{
    public string AnimalType { get; set; }
    public bool IsDomestic { get; set; }
    
    public Mammal(int age, string species, string animalType, bool isDomestic) 
        : base(age, species)
    {
        AnimalType = animalType;
        IsDomestic = isDomestic;
    }
    
    public override void Eat()
    {
        Console.WriteLine($"üåø {AnimalType} –ø–∏—Ç–∞–µ—Ç—Å—è —Ç—Ä–∞–≤–æ–π –∏ —Ä–∞—Å—Ç–µ–Ω–∏—è–º–∏");
    }
    
    public override string ToString()
    {
        string domesticStatus = IsDomestic ? "–¥–æ–º–∞—à–Ω–µ–µ" : "–¥–∏–∫–æ–µ";
        return $"–ú–ª–µ–∫–æ–ø–∏—Ç–∞—é—â–µ–µ: {AnimalType}, {base.ToString()}, –°—Ç–∞—Ç—É—Å: {domesticStatus}";
    }
}

public class Predator : Animal
{
    public bool IsDomestic { get; set; }
    public string AnimalType { get; set; }
    
    public Predator(int age, string species, string animalType, bool isDomestic) 
        : base(age, species)
    {
        AnimalType = animalType;
        IsDomestic = isDomestic;
    }
    
    public override void Eat()
    {
        Console.WriteLine($"ü•© {AnimalType} –æ—Ö–æ—Ç–∏—Ç—Å—è –∏ –ø–∏—Ç–∞–µ—Ç—Å—è –º—è—Å–æ–º");
    }
    
    public override string ToString()
    {
        string domesticStatus = IsDomestic ? "–¥–æ–º–∞—à–Ω–µ–µ" : "–¥–∏–∫–æ–µ";
        return $"–•–∏—â–Ω–∏–∫: {AnimalType}, {base.ToString()}, –°—Ç–∞—Ç—É—Å: {domesticStatus}";
    }
}

class Program
{
    static void Main()
    {
        Console.WriteLine("ü¶Å –°–∏—Å—Ç–µ–º–∞ —É—á–µ—Ç–∞ –∂–∏–≤–æ—Ç–Ω—ã—Ö\n");
        
        // –ü—Ä–∏–º–µ—Ä—ã –º–ª–µ–∫–æ–ø–∏—Ç–∞—é—â–∏—Ö
        Mammal sheep = new Mammal(5, "–ú–ª–µ–∫–æ–ø–∏—Ç–∞—é—â–µ–µ", "–û–≤–µ—á–∫–∞", true);
        Mammal cow = new Mammal(4, "–ú–ª–µ–∫–æ–ø–∏—Ç–∞—é—â–µ–µ", "–ö–æ—Ä–æ–≤–∞", true);
        Mammal deer = new Mammal(3, "–ú–ª–µ–∫–æ–ø–∏—Ç–∞—é—â–µ–µ", "–û–ª–µ–Ω—å", false);
        
        // –ü—Ä–∏–º–µ—Ä—ã —Ö–∏—â–Ω–∏–∫–æ–≤
        Predator wolf = new Predator(7, "–•–∏—â–Ω–∏–∫", "–í–æ–ª–∫", false);
        Predator tiger = new Predator(3, "–•–∏—â–Ω–∏–∫", "–¢–∏–≥—Ä", false);
        Predator cat = new Predator(2, "–•–∏—â–Ω–∏–∫", "–ö–æ—à–∫–∞", true);

        // –í—ã–≤–æ–¥–∏–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –∂–∏–≤–æ—Ç–Ω—ã—Ö
        Console.WriteLine("üìã –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –º–ª–µ–∫–æ–ø–∏—Ç–∞—é—â–∏—Ö:");
        Console.WriteLine(sheep);
        Console.WriteLine(cow);
        Console.WriteLine(deer);
        
        Console.WriteLine("\nüìã –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Ö–∏—â–Ω–∏–∫–∞—Ö:");
        Console.WriteLine(wolf);
        Console.WriteLine(tiger);
        Console.WriteLine(cat);
        
        Console.WriteLine($"\nüìä –í—Å–µ–≥–æ —Å–æ–∑–¥–∞–Ω–æ –∂–∏–≤–æ—Ç–Ω—ã—Ö: {Animal.AnimalCount}");
        
        Console.WriteLine("\nüçΩÔ∏è  –ü—Ä–æ—Ü–µ—Å—Å –ø–∏—Ç–∞–Ω–∏—è:");
        sheep.Eat();
        cow.Eat();
        deer.Eat();
        wolf.Eat();
        tiger.Eat();
        cat.Eat();
        
        Console.WriteLine($"\nüéØ –•–∏—â–Ω–∏–∫–æ–≤ —Å–æ–∑–¥–∞–Ω–æ: 3");
        Console.WriteLine($"üåø –ú–ª–µ–∫–æ–ø–∏—Ç–∞—é—â–∏—Ö —Å–æ–∑–¥–∞–Ω–æ: 3");
        
        // –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –ø—Ä–∏–º–µ—Ä—ã
        Console.WriteLine("\n‚ûï –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –ø—Ä–∏–º–µ—Ä—ã:");
        Mammal goat = new Mammal(2, "–ú–ª–µ–∫–æ–ø–∏—Ç–∞—é—â–µ–µ", "–ö–æ–∑–∞", true);
        Predator dog = new Predator(5, "–•–∏—â–Ω–∏–∫", "–°–æ–±–∞–∫–∞", true);
        
        Console.WriteLine(goat);
        Console.WriteLine(dog);
        goat.Eat();
        dog.Eat();
        
        Console.WriteLine($"\nüìä –ò—Ç–æ–≥–æ –∂–∏–≤–æ—Ç–Ω—ã—Ö: {Animal.AnimalCount}");
    }
}
````
